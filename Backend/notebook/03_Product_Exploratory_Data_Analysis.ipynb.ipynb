```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", 500)
pd.set_option("display.width", 1200)

plt.style.use("ggplot")


print("Libraries Loaded Successfully")
```

```python
DATA_PATH = "../processed/amazon_products_unique.csv"

products = pd.read_csv(DATA_PATH)

print("Dataset Loaded Successfully")
```

```python
rows, cols = products.shape

print(f"Rows    : {rows}")
print(f"Columns : {cols}")
```

```python
products.describe().T
```

```python
missing = pd.DataFrame({

    "Missing Values": products.isnull().sum(),

    "Percentage": (
        products.isnull().mean()*100
    ).round(2)

})

missing
```

```python
plt.figure(figsize=(12,5))

sns.barplot(

    x=missing.index,

    y=missing["Percentage"]

)

plt.xticks(rotation=90)

plt.ylabel("Missing (%)")

plt.title("Missing Value Analysis")

plt.show()
```

```python
duplicate_products = products.duplicated().sum()

print("Duplicate Rows :", duplicate_products)
```

```python
print("Unique Product IDs :")

print(products["product_id"].nunique())
```

```python
quality = pd.DataFrame({

    "Data Type": products.dtypes,

    "Missing": products.isnull().sum(),

    "Unique Values": products.nunique()

})

quality
```

```python
numeric_columns = products.select_dtypes(
    include=["number"]
).columns

print("Numeric Features")

numeric_columns
```

```python
categorical_columns = products.select_dtypes(
    exclude=["number"]
).columns

print("Categorical Features")

categorical_columns
```

```python
report = {

    "Total Products":
        len(products),

    "Unique Products":
        products["product_id"].nunique(),

    "Average Rating":
        round(products["rating"].mean(),2),

    "Average Price":
        round(products["discounted_price"].mean(),2),

    "Highest Price":
        products["discounted_price"].max(),

    "Lowest Price":
        products["discounted_price"].min()

}

report
```

```python
for key, value in report.items():

    print(f"{key:<25}: {value}")
```

```python
price_stats = products[
    [
        "discounted_price",
        "actual_price"
    ]
].describe().round(2)

price_stats
```

```python
plt.figure(figsize=(10,6))

plt.hist(
    products["discounted_price"],
    bins=40,
    edgecolor="black"
)

plt.title("Discounted Price Distribution")

plt.xlabel("Price")

plt.ylabel("Products")

plt.show()
```

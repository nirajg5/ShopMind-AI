

## Product & Review Dataset Creation



```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", 300)

print("Libraries Loaded Successfully")
```

```python
df = pd.read_csv("../processed/amazon_products_cleaned.csv")

print("Dataset Loaded")
```

```python
df.head(2)
```

```python
print(df.shape)
```

```python
duplicate_products = df["product_id"].duplicated().sum()

print("Duplicate Product IDs :", duplicate_products)
```

```python
print("Total Rows :", len(df))

print("Unique Products :", df["product_id"].nunique())

print("Total Reviews :", len(df))
```

```python
review_count = (
    df.groupby("product_id")
      .size()
      .sort_values(ascending=False)
)

review_count.head(10)
```

```python
products = (
    df
    .sort_values("product_id")
    .drop_duplicates(
        subset="product_id",
        keep="first"
    )
    .reset_index(drop=True)
)
```

```python
products.shape
```

```python
products.head(2)
```

```python
print(df.columns)
```

```python
reviews = df[
    [
        "product_id",
        "product_name",
        "review_title",
        "review_content",
        "rating",
        "rating_count",
        "review_length",
        "review_word_count"
    ]
].copy()

reviews.head()
```

```python
print("Review Dataset Shape")
print(reviews.shape)
```

```python
print("Total Reviews :", len(reviews))

print("Unique Products :", reviews["product_id"].nunique())
```

```python
reviews[
    [
        "review_title",
        "review_length",
        "review_word_count"
    ]
].head()
```

```python
avg_rating = (
    reviews
    .groupby("product_id")["rating"]
    .mean()
    .reset_index()
)

avg_rating.rename(
    columns={
        "rating":"average_rating"
    },
    inplace=True
)

avg_rating.head()
```

```python
products = products.merge(
    avg_rating,
    on="product_id",
    how="left"
)

products.head()
```

```python
review_count = (
    reviews
    .groupby("product_id")
    .size()
    .reset_index(name="total_reviews")
)

review_count.head()
```

```python
products = products.merge(
    review_count,
    on="product_id",
    how="left"
)

products.head()
```

```python
products[
    [
        "product_name",
        "average_rating",
        "total_reviews"
    ]
].head()
```

```python
products.isnull().sum()
```

```python
reviews.isnull().sum()
```

```python
memory = (
    products.memory_usage(deep=True).sum()
    /
    (1024*1024)
)

print(f"Memory Usage : {memory:.2f} MB")
```

```python
products.to_csv(
    "../processed/amazon_products_unique.csv",
    index=False
)

print("Products Dataset Saved")
```

```python
reviews.to_csv(
    "../processed/amazon_reviews.csv",
    index=False
)

print("Review Dataset Saved")
```

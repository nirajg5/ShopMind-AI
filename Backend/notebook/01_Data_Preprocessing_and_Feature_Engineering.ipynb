

## Data Cleaning


```python
df_clean = df.copy()

print("Working Copy Created Successfully")
```

```python
def clean_price(price):

    if pd.isna(price):
        return np.nan

    price = str(price)

    price = price.replace("₹", "")

    price = price.replace(",", "")

    price = price.strip()

    try:
        return float(price)

    except:
        return np.nan
```

```python
df_clean["discounted_price"] = (
    df_clean["discounted_price"]
    .apply(clean_price)
)

df_clean["actual_price"] = (
    df_clean["actual_price"]
    .apply(clean_price)
)
```

```python
df_clean[
    [
        "discounted_price",
        "actual_price"
    ]
].head()
```

```python
def clean_discount(discount):

    if pd.isna(discount):
        return np.nan

    discount = str(discount)

    discount = discount.replace("%", "")

    discount = discount.strip()

    try:
        return float(discount)

    except:
        return np.nan
```

```python
df_clean["discount_percentage"] = (
    df_clean["discount_percentage"]
    .apply(clean_discount)
)
```

```python
df_clean["discount_percentage"].head()
```

```python
def clean_rating(rating):

    if pd.isna(rating):
        return np.nan

    rating = str(rating).strip()

    try:
        return float(rating.split()[0])

    except:
        return np.nan
```

```python
df_clean["rating"] = (
    df_clean["rating"]
    .apply(clean_rating)
)
```

```python
df_clean[
    df_clean["rating"].isna()
][
    [
        "product_name",
        "rating"
    ]
]
```

```python
median_rating = df_clean["rating"].median()

df_clean["rating"] = (
    df_clean["rating"]
    .fillna(median_rating)
)
```

```python
def clean_rating_count(count):

    if pd.isna(count):
        return 0

    count = str(count)

    count = count.replace(",", "")

    count = count.strip()

    try:
        return int(float(count))

    except:
        return 0
```

```python
df_clean["rating_count"] = (
    df_clean["rating_count"]
    .apply(clean_rating_count)
)
```

```python
df_clean[
    [
        "rating",
        "rating_count"
    ]
].head()
```

```python
df_clean.isnull().sum()
```

```python
df_clean["category_list"] = (
    df_clean["category"]
    .str.split("|")
)
```

```python
df_clean["num_categories"] = (
    df_clean["category_list"]
    .apply(len)
)
```

```python
# Columns to remove
columns_to_drop = [
    "category",
    "user_id",
    "user_name",
    "review_id",
    "product_link"
]

# Drop columns
df_clean = df_clean.drop(columns=columns_to_drop, errors="ignore")

# Verify
print("Remaining Columns:")
print(df_clean.columns.tolist())
```



# Feature Engineering


```python
df_clean["discount_amount"] = (
    df_clean["actual_price"]
    - df_clean["discounted_price"]
)
```

```python
df_clean[
    [
        "actual_price",
        "discounted_price",
        "discount_amount"
    ]
].head()
```

```python
bins = [
    0,
    1000,
    5000,
    10000,
    50000,
    100000,
    np.inf
]

labels = [
    "Very Cheap",
    "Cheap",
    "Budget",
    "Mid Range",
    "Premium",
    "Luxury"
]

df_clean["price_bucket"] = pd.cut(
    df_clean["discounted_price"],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
df_clean["price_bucket"].value_counts()
```

```python
df_clean["review_length"] = (
    df_clean["review_content"]
    .fillna("")
    .str.len()
)
```

```python
df_clean["description_length"] = (
    df_clean["about_product"]
    .fillna("")
    .str.len()
)
```

```python
df_clean["description_length"].head()
```

```python
df_clean["product_name_length"] = (
    df_clean["product_name"]
    .fillna("")
    .str.len()
)
```

```python
df_clean["review_word_count"] = (
    df_clean["review_content"]
    .fillna("")
    .str.split()
    .str.len()
)
```

```python
df_clean["description_word_count"] = (
    df_clean["about_product"]
    .fillna("")
    .str.split()
    .str.len()
)
```

```python
df_clean["popularity_score"] = (
    df_clean["rating"]
    *
    np.log1p(df_clean["rating_count"])
)
```

```python
df_clean["popularity_score"] = (
    df_clean["rating"]
    *
    np.log1p(df_clean["rating_count"])
)
```

```python
df_clean["popularity_score"].head()
```

```python
df_clean["price_per_rating"] = (
    df_clean["discounted_price"]
    /
    df_clean["rating"]
)
```

```python
df_clean["price_per_rating_inf"] = (
    df_clean["price_per_rating"]
    .replace(
        [np.inf, -np.inf],
        np.nan
    )
)
```

```python
def build_search_document(row):

    return f"""
Product Name:
{row['product_name']}

Product Description:
{row['about_product']}

Customer Rating:
{row['rating']}

Number of Ratings:
{row['rating_count']}

Discount Price:
₹{row['discounted_price']}

Actual Price:
₹{row['actual_price']}

Discount:
{row['discount_percentage']}%

Review Title:
{row['review_title']}

Customer Review:
{row['review_content']}
"""
```

```python
df_clean["search_document"] = (
    df_clean.apply(
        build_search_document,
        axis=1
    )
)
```

```python
category_count = (
    df["category_list"]
    .value_counts()
    .head(15)
)

category_count
```

```python
plt.figure(figsize=(14,6))

sns.barplot(
    x=category_count.values,
    y=category_count.index
)

plt.title("Top Categories")

plt.xlabel("Number of Products")

plt.ylabel("Category")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["discounted_price"],
    bins=30,
    kde=True
)

plt.title("Discounted Price Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["actual_price"],
    bins=30,
    kde=True
)

plt.title("Actual Price Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["rating"],
    bins=20,
    kde=True
)

plt.title("Rating Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    np.log1p(df["rating_count"]),
    bins=30
)

plt.title("Log Rating Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["discount_percentage"],
    bins=25
)

plt.title("Discount Distribution")

plt.show()
```

```python
import numpy as np
import pandas as pd

bins = [
    0,
    1000,
    5000,
    10000,
    50000,
    100000,
    np.inf
]

labels = [
    "Very Cheap",
    "Cheap",
    "Budget",
    "Mid Range",
    "Premium",
    "Luxury"
]

df["price_bucket"] = pd.cut(
    df["discounted_price"],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```
